Variables A and B are not conditionally independent given D, F, and G. This is because there are several unblocked trails connecting A to B:

A ‚Äì C ‚Äì F: a causal trail that remains active since C is not observed.
C ‚Äì F ‚Äì E: a common effect trail that becomes active due to the observation of F.
F ‚Äì E ‚Äì B: a common cause trail that is active since E is not observed.
Together, these trails create an open path from A to B, so conditional independence does not hold.

B.
Any trail connecting D and G must pass through F. Examples include:

A ‚Äì F ‚Äì G,

B ‚Äì F ‚Äì G (common cause trail).
Since observing F blocks both causal and common cause trails, D and G are conditionally independent given F.

A ‚Äì F ‚Äì B,

**QUESTION 5** [10pts]

**A.** Given a random variable $X \sim \mathrm{Bernoulli}(\theta)$. That is, $P(X = 1) = \theta$ and $P(X = 0) = 1 - \theta$.

Suppose we are given $M + N$ samples of $X$ such that $X = 1$ in M samples and $X = 0$ in the remaining $N$ samples. Find the maximum likelihood estimate (MLE) of $\theta$ as a function of $M$ and $N$. [5pts]

**B.** Given a random variable $X \sim \mathrm{Cat}(1 - \theta, \theta/2, \theta/2)$ where $\theta \in (0, 1)$.

Suppose we are given $N_0 + N_1 + N_2$ independent samples of $X$ such that in $N_0$ samples, $X = 0$. In $N_1$ samples, $X = 1$. In the remaining $N_2$ samples, $X = 2$. Find the maximum likelihood estimate (MLE) of $\theta$ as a function of $N_0$, $N_1$, and $N_2$. [5pts]

**SOLUTION**

**A.** Since we are given  $P(X = 1) = \theta$ and $P(X = 0) = 1 - \theta$, the chancethat we get $M$ samples of $X = 1$ and $N$ samples of $X = 0$ is:

$\theta^M \cdot (1 - \theta)^N$

The max estimation of $\theta$ maximizes the expression above, or equivalently, its logarithmic form,

$L(\theta) = M\log\theta + N\log(1 - \theta)$

Setting $\mathrm{d}L/\mathrm{d}\theta = 0$ and solving for $\theta$:

$\mathrm{d}L/\mathrm{d}\theta = M/\theta - N/(1 - \theta) = 0$ which implies $\theta = M/(M + N)$.

**B.** In this case, $X$ can take $3$ values (i.e., 0, 1, and 2) with the following probabilities:

$P(X = 0) =  1 - \theta$

$P(X = 1) = \theta/2$

$P(X = 2) = \theta/2$

So in the end the probability of drawing $N_0$ samples of $X = 0$, $N_1$ samples $X = 1$, and $N_2$ samples $X = 2$ is:

$(1 - \theta)^{N_0} \cdot (\theta/2)^{N_1} \cdot (\theta/2)^{N_2} = (1 - \theta)^{N_0} \cdot (\theta/2)^{N_1 + N_2}$

Similar to part A, we would want to find the value of $\theta$ that maximizes the expression above, or equivalently, its logarithmic form:

$L(\theta) = N_0\log(1 - \theta) + (N_1 + N_2) \log (\theta/2)$

Setting $\mathrm{d}L/\mathrm{d}\theta = 0$ and solving for $\theta$:

$\mathrm{d}L/\mathrm{d}\theta = -N_0 / (1 - \theta) + (N_1 + N_2) / \theta = 0$

which implies $\theta = 1 - (N_0 / (N_0 + N_1 + N_2))$.
