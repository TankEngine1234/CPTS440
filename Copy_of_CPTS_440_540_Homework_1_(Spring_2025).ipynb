h(n) ≤ h*(n), where h*(n) is the true cost to the goal.

h1(n) ≤ h*(n)

**A.** Work out an implementation of the breadth-first-search algorithm that find all the nodes that can be reached from an initial node $x$ either directly or via a path of directed edges. In particular, fill in the code template below to substantiate the algorithm. [20pts]

```python
from typing import List, Dict, Set

def initialize(x: int): # initialize the frontier/explored tables given X
    frontier = [x]
    explored = set()
    return frontier, explored

def select(frontier: List[int]) -> int: # remove and return a node from the frontier -- this node will be explored next
    return frontier.pop(0)

def explore(node: int, A: Dict[int, List[int]], explored: Set[int]) -> List[int]: # generate a list Q of nodes that 'node' can reach via a direct edge on the graph A -- ignore nodes that are in either Frontier or Explored
    Q = []
    for neighbor in A[node]:
        if neighbor not in explored:
            Q.append(neighbor)
    return Q

def update(frontier: List[int], explored: Set[int], eligibles: List[int], current_node: int) -> tuple[List[int], Set[int]]: # given a list of 'eligibles' (i.e., nodes that can be added to the Frontier), update the Frontier and Explored list accordingly
    # Add the current node to explored, not the last frontier node
    explored.add(current_node)

    for node in eligibles:
        if node not in frontier and node not in explored:
            frontier.append(node)
    return frontier, explored

def search(A: Dict[int, List[int]], x: int) -> List[int]:
    # initialize the frontier and explored tables
    frontier, explored = initialize(x) # initialize frontier and explored tables
    res = [] # this is the list of reachable nodes

    while len(frontier) > 0:
        node = select(frontier) # select a node to be explored next
        res.append(node) # this means we are able to reach 'node' from x, so we put it in x's list of reachable nodes
        eligibles = explore(node, A, explored) # discover all nodes that (1) can be reached from the selected node above, and (2) have never been on the frontier/explored list
        frontier, explored = update(frontier, explored, eligibles, node) # update the frontier and explored list with the eligibles

    return res # return result: a list of nodes which are reachable from x

def test_bfs():
    # Test Case 1: The example given in the problem
    A1 = {0: [1], 1: [2, 3], 2: [3, 4], 3: [], 4: []}
    print("Test Case 1:")
    print("Graph:", A1)
    print("Starting from node 0:", search(A1, 0))  # Should print [0, 1, 2, 3, 4]
    print("Starting from node 1:", search(A1, 1))  # Should print [1, 2, 3, 4]
    print("Starting from node 3:", search(A1, 3))  # Should print [3]

    # Test Case 2: Cyclic graph
    A2 = {0: [1], 1: [2], 2: [0, 3], 3: []}
    print("\nTest Case 2 (Cyclic graph):")
    print("Graph:", A2)
    print("Starting from node 0:", search(A2, 0))  # Should print [0, 1, 2, 3]

    # Test Case 3: Disconnected graph
    A3 = {0: [1], 1: [], 2: [3], 3: [2]}
    print("\nTest Case 3 (Disconnected graph):")
    print("Graph:", A3)
    print("Starting from node 0:", search(A3, 0))  # Should print [0, 1]
    print("Starting from node 2:", search(A3, 2))  # Should print [2, 3]

    # Test Case 4: Single node
    A4 = {0: []}
    print("\nTest Case 4 (Single node):")
    print("Graph:", A4)
    print("Starting from node 0:", search(A4, 0))  # Should print [0]

    # Test Case 5: Complete graph
    A5 = {0: [1, 2], 1: [0, 2], 2: [0, 1]}
    print("\nTest Case 5 (Complete graph):")
    print("Graph:", A5)
    print("Starting from node 0:", search(A5, 0))  # Should print [0, 1, 2]

test_bfs()
```

```python
from collections import deque

def initialize(x):
    frontier = deque([(x, 0)])  # queue for BFS
    explored = set([x])  # visited states
    return frontier, explored

def select(frontier):
    return frontier.popleft()  # dequeue for BFS

def find_next(state):
    states = []
    idx = state.index('*')  # Find the blank space
    row, col = divmod(idx, 3)  # Convert index to row, col
    moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]  # Up, Down, Left, Right

    for dr, dc in moves:
        new_row, new_col = row + dr, col + dc
        if 0 <= new_row < 3 and 0 <= new_col < 3:
            new_idx = new_row * 3 + new_col
            new_state = list(state)
            new_state[idx], new_state[new_idx] = new_state[new_idx], new_state[idx]
            states.append("".join(new_state))

    return states

def expand(frontier, explored, state, cost):
    for next_state in find_next(state):
        if next_state not in explored:
            frontier.append((next_state, cost + 1))
            explored.add(next_state)
    return frontier, explored

def minstep(x) -> int:
    frontier, explored = initialize(x)
    solution = '12345678*'

    while frontier:
        state, cost = select(frontier)
        if state == solution:
            return cost
        frontier, explored = expand(frontier, explored, state, cost)

    return -1  # No solution found

# Test cases
print(minstep('1234*5786'))  # Expected output: 2
print(minstep('15243*786'))  # Example input, should return the minimum swaps
print(minstep('12345678*'))  # Already solved, should return 0
print(minstep('123456*78'))
print(minstep('1*2345678'))
print(minstep('123*45678'))
print(minstep('21345678*'))  #unsolvable output:-1
```
